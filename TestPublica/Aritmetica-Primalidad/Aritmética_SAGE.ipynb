```python

bits=1024
bits_p=10
for bits in [2**i for i in range(10,20)]:
    p=randint(1,2**(bits_p))
    n1=randint(1,2**(bits-bits_p))*p
    n2=randint(1,2**(bits-bits_p))*p
    %time bezout=xgcd(n1,n2)
    print("\nBits: {0}, Cifras decimales: ~{1}".format(bits,int(log(min(n1,n2),10))))
    display(Latex("$n_1$= {0} \n\n $n_2$={1} \n\n mcd$(n_1,n_2)$={2}".format(n1, n2, bezout[0])))

```
